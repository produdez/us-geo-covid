## Loading real data
`Note: must run server to request for data`

```python
import requests
import json
def get(path):
    result = requests.get(f'http://127.0.0.1:8000/api-covid/{path}/')._content
    result = json.loads(result)
    return result
```

```python
state = get('state/name/Utah')
id = state['id']
print(f'State name: {state["name"]}, state_id: {id}')
state_data = get(f'state-reports/{id}')['results']
print('report count: ', len(state_data))
print('data example: ', state_data[0])
```

# Let's go

```python
display(Javascript(filename="./state-line-graph/line-graph.js"))
def draw_line_graph(data, width=600, height=400):
    display(Javascript("""
        (function(element){
            require(['line_graph'], function(line_graph) {
                line_graph(element.get(0), %s);
            });
        })(element);
    """ % (json.dumps(data))))
draw_line_graph(state_data)
```

# Pie chart

```python
# get us global data (daily)
date_str = '2020/10/04'
today_data = get(f'date-reports/{date_str}')['results']
all_states_info = get(f'states')['results']
print('report count: ', len(today_data))
print('data example: ', today_data[0])
print('first state: ', all_states_info[0])
date = today_data[0]['date']
for report in today_data:
    if report['date'] != date: print('Error, reports not same date')
```

```python
display(Javascript(filename="./us-waffle-chart/waffle-chart.js"))
def draw_waffle_chart(data_today, data_states, width=600, height=400):
    display(Javascript("""
        (function(element){
            require(['waffle_chart'], function(waffle_chart) {
            
                waffle_chart(element.get(0), %s, %s);
            });
        })(element);
    """ % (json.dumps(data_today), json.dumps(data_states))))
draw_waffle_chart(today_data, all_states_info)
```
